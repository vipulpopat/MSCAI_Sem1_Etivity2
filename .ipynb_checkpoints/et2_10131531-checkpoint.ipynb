```python
# There's missing values for education, see what the distribution is and fill in the missing value with the most popular as a best guess to approximate the age.
df['education'].value_counts()
df['education'].fillna('secondary', inplace=True)
```

```python
table = df.pivot_table(values='age', 
                       index='education', 
                       columns=['marital'], 
                       aggfunc=np.median)
print(table)
```

```python
# Define function to return an element of the pivot table
def get_element(x):
    return table.loc[x['education'], x['marital']]

# Replace missing values
df['age'].fillna(df[df['age'].isnull()].apply(get_element, axis=1), inplace=True)
```

```python
# Check that job and education have been filled
df.apply(lambda x: sum(x.isnull()), axis=0)
```

we can see age and education are now 0, so those missing values have been filled in.

```python
df['job'].fillna('unknown', inplace=True)
df['contact'].fillna('unknown', inplace=True)
df['poutcome'].fillna('unknown', inplace=True)
```

```python
Eins = []
Eouts = []

def run_experiment(counter, X_train, X_test, y_train, y_test):
    for i in range (counter):
        X_train, X_test, y_train, y_test = train_test_split(X_shuffled, y_shuffled, test_size=0.2)

        Ein = 1 - logistic_classifer.score(X_train, y_train)
        Eout = 1 - logistic_classifer.score(X_test, y_test)
        Eins.append(Ein)
        Eouts.append(Eout)

def calc_prob_epsilon(M, Eins, Eouts, epsilons):
    prob_epsilon=[]
    for e in epsilons:   
        counter = 0
        for i in range(M):
            if abs(Eins[i] - Eouts[i])>e:
                   counter +=1
        prob_epsilon.append(counter/M)           
    return prob_epsilon


run_experiment(1000, X_train, X_test, y_train, y_test)
```

```python
run_experiment(1000, X2_train, X2_test, y2_train, y2_test)

epsilons = np.arange(0, 1, 0.01)
error_in = [Ein]
error_out = [Eout]
prob_epsilon = calc_prob_epsilon(1, Eins, Eouts, epsilons)

print("epsilon proba. dist:", prob_epsilon)
```

```python
epsilons = np.arange(0, 1, 0.01)
N=2000
```

```python
import math
tolerance = 0.01
dvc = 3
def get_vc_bound(N, dvc, tolerance, Ein):
    return Ein + math.sqrt((1/(2*N)) * math.log((2 * (math.pow(N,dvc) + 1)/tolerance) ))

print(get_vc_bound(N, dvc, tolerance, Ein))
```

```python

error_diff = abs(np.array(error_in)-np.array(error_out))
epsilons = np.arange(0.001,0.20,0.001)
prob_epsilon=[]
for e in epsilons:
    prob_epsilon.append(len([1 for i in error_diff if i > e])/len(error_diff))
plt.plot(epsilons, prob_epsilon, label='P[|Ein-Eout|>e]')

hoef = [2*math.exp(-2*N*e**2) for e in epsilons]
vc = [4*((2*N)**3+1)*math.exp(-(N/8)*e**2) for e in epsilons]

print("Error exceeds Hoeffding" if max(np.array(prob_epsilon)-np.array(hoef))>0 else "Error does not exceed Hoeffding")
print("Error exceeds VC" if max(np.array(prob_epsilon)-np.array(vc))>0 else "Error does not exceed VC")
plt.plot(epsilons, hoef,'r', label='Hoeffding Bound')
plt.plot(epsilons, vc, 'g', label='VC Bound')
plt.yscale('log')
plt.legend(loc='upper right')
plt.show()
```

## Task 5 (CE6002)
Using the dataset, make learning curves for logistic regression and two other machine learning algorithms of your choosing using scikitlearn’s learning_curve functionality. Choose one algorithm with low complexity in terms of VC dimension and one algorithm with high complexity.

Comment on what information you can derive from the learning curves  (in a markdown cell) and how this information differs between simple and complex machine learning algorithms.

```python
def plot_learning_curve(estimator):
    cv = ShuffleSplit(n_splits=100, test_size=0.2, random_state=0)
    n_jobs=None
    
    # Create train sizes
    train_sizes=np.linspace(0.1, 1.0, 5)
    print("Train sizes selected :\n", train_sizes)


    # Determine the cross-validated training and test scores for different training set sizes.
    train_sizes, train_scores, test_scores = learning_curve(estimator=estimator, 
                                                            X=X_train, 
                                                            y=y_train, 
                                                            cv=cv, 
                                                            n_jobs=n_jobs, 
                                                            train_sizes=train_sizes)
#     Printing all the values just adds noise, 
#     uncomment if you want to debug or look at values
#     print("train_scores:\n", train_scores)
#     print("test_scores :\n", test_scores)

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std  = np.std(train_scores, axis=1)
    test_scores_mean  = np.mean(test_scores, axis=1)
    test_scores_std   = np.std(test_scores, axis=1)

    label_train = "Training " + str(estimator)[:10] + "..."
    label_cross_valid = "Cross-valid. score " + str(estimator)[:10] + "..."
  
    plt.plot(train_sizes, train_scores_mean, 'ro--', label=label_train)
    plt.plot(train_sizes, test_scores_mean, 'go--', label=label_cross_valid)
    
    # Highlight the standtard deviation from mean
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std, 
                     train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    
    plt.legend(loc="best")
    plt.ylabel('Score')
    plt.xlabel('Training examples')
    
    plt.grid()
    plt.ylim(0.3, 1.01)
    plt.show()
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.ensemble import AdaBoostClassifier

estimators = [LogisticRegression(), SVC(), GaussianNB(), AdaBoostClassifier()]

fig = plt.figure(figsize=(15,20))
counter = 0
for e in estimators:
    counter +=1
    ax2 = fig.add_subplot(5, 2, counter)
    ax2.set_xlabel('Train size')
    ax2.set_ylabel('Score')
    plot_learning_curve(e)
plt.tight_layout()
```

```python
from sklearn import metrics

def plot_confusion(cm_result, labels , title):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(cm_result)
    plt.title(title)
    fig.colorbar(cax)
    ax.set_xticklabels(labels)
    ax.set_yticklabels(labels)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
```

```python
# SVC 
```

```python
svclassifier = SVC() 
svclassifier.fit(X_train, y_train) 
y_pred = svclassifier.predict(X_test) 
print('Accurary', metrics.accuracy_score(y_test, y_pred))
cm_result = metrics.confusion_matrix(y_test, y_pred)
plot_confusion(cm_result,y_pred, 'SVC Confusion Matrix')

Ein = 1 - svclassifier.score(X_train, y_train)
print("Ein:", Ein)

Eout = 1 - svclassifier.score(X_test, y_test)
print("Eout:", Eout)
```

```python
# GaussianNB(), 
```

```python
gaussian_classifier = GaussianNB()
gaussian_classifier.fit(X_train, y_train)
y_gaussian = gaussian_classifier.predict(X_test) 
print('Accurary', metrics.accuracy_score(y_test, y_gaussian))
ada_cm = metrics.confusion_matrix(y_test, y_gaussian)
plot_confusion(ada_cm, y_gaussian, 'Gaussian Confusion Matrix')

Ein = 1 - gaussian_classifier.score(X_train, y_train)
print("Ein:", Ein)

Eout = 1 - gaussian_classifier.score(X_test, y_test)
print("Eout:", Eout)
```

```python
adaboost_classifier = AdaBoostClassifier()
adaboost_classifier.fit(X_train, y_train)
y_adaboostpred = adaboost_classifier.predict(X_test) 
print('Accuracy', metrics.accuracy_score(y_test, y_adaboostpred))
ada_cm = metrics.confusion_matrix(y_test, y_adaboostpred)
plot_confusion(ada_cm,y_adaboostpred, 'Adaboost Confusion Matrix')

Ein = 1 - adaboost_classifier.score(X_train, y_train)
print("Ein:", Ein)

Eout = 1 - adaboost_classifier.score(X_test, y_test)
print("Eout:", Eout)
```
